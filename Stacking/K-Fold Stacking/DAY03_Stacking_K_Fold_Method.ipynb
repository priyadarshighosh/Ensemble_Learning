# Training

```python
df = pd.read_csv('heart.csv')
X = df.drop('target', axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

# Different Stacking Methods

```python
base_models = [
    ('rf', RandomForestClassifier(random_state=42)),
    ('gb', GradientBoostingClassifier(random_state=42)),
    ('svm', SVC(probability=True, random_state=42))
]


meta_learner = LogisticRegression(random_state=42)        # This is the last layer
```

## 3. K-Fold Stacking

In this method we Divide the training dataset into 4 parts k1 , k2 ,k3 ,kk4 and we will train 1st model with k1 k2 k3 and predict on k4 and then k1 k2 k4 and predict on k3 Like this we will have prediction on all ks . This process will take place for all the models in the 1st layer .

After getting the output we will train the Meta Model that is the Final Layer model

In this approach we first train the Meta Model then we train the Base model so after training the meta model we will put the entire dataset into the 3 base models and train

After training the base and the meta model we will pass the  test dataset through it to get the output .

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, shuffle=True, random_state=42)   # no of k fold

meta_train_kfold = pd.DataFrame(index=X_train.index)    # enpty set where we will store the result from the base-models
meta_test_kfold = pd.DataFrame(index=X_test.index)

for name, model in base_models:
    meta_train_kfold[name] = 0
    meta_test_kfold[name] = 0


    for train_index, val_index in kf.split(X_train):
        X_train_kf, X_val_kf = X_train.iloc[train_index], X_train.iloc[val_index]
        y_train_kf, y_val_kf = y_train.iloc[train_index], y_train.iloc[val_index]

        model.fit(X_train_kf, y_train_kf)

        meta_train_kfold.iloc[val_index, meta_train_kfold.columns.get_loc(name)] = model.predict_proba(X_val_kf)[:, 1]
        meta_test_kfold[name] += model.predict_proba(X_test)[:, 1] / 5




meta_learner.fit(meta_train_kfold, y_train)             # the last layer is fitted with the training data from the Other layer
y_pred_kfold = meta_learner.predict(meta_test_kfold)


accuracy_kfold = accuracy_score(y_test, y_pred_kfold)       # calculates and accuracy score
print(f"Stacking K-Fold Accuracy: {accuracy_kfold}")
```
